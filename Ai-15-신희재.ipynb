```python
df = pd.read_csv('vgames2.csv')
```

```python
df = df.dropna()
```

```python
def change(a):
    change_list = []

    for i in a:
        temp = float(i.replace("K",""))
        change_list.append(temp)

    return change_list
```

```python
 les_X = df.loc[df.NA_Sales.str.contains('K')]
 les_X['X'] = change(les_X['NA_Sales'])
 les_X['X'] = les_X['X']*1000
 f = les_X['X'].index
```

```python
df['NA_Sales'] = df['NA_Sales'].str.replace("K", "");
df['NA_Sales'] = df['NA_Sales'].str.replace("M", "");
df['NA_Sales'] = df['NA_Sales'].astype(float)
df['NA_Sales'] = df['NA_Sales']*1000000
```

```python
df['NA_Sales'].loc[f] = les_X['X']
df['NA_Sales'] = df['NA_Sales'].astype(int)
```

```python
 les_Y = df.loc[df.EU_Sales.str.contains('K')]
 les_Y['Y'] = change(les_Y['EU_Sales'])
 les_Y['Y'] = les_Y['Y']*1000
 j = les_Y['Y'].index
 
```

```python
df['EU_Sales'] = df['EU_Sales'].str.replace("K", "");
df['EU_Sales'] = df['EU_Sales'].str.replace("M", "");
df['EU_Sales'] = df['EU_Sales'].astype(float)
df['EU_Sales'] = df['EU_Sales']*1000000
```

```python
df['EU_Sales'].loc[j] = les_Y['Y']
df['EU_Sales'] = df['EU_Sales'].astype(int)
```

```python
 les_Z = df.loc[df.JP_Sales.str.contains('K')]
 les_Z['Z'] = change(les_Z['JP_Sales'])
 les_Z['Z'] = les_Z['Z']*1000
 t = les_Z['Z'].index;
 
```

```python
df['JP_Sales'] = df['JP_Sales'].str.replace("K", "");
df['JP_Sales'] = df['JP_Sales'].str.replace("M", "");
df['JP_Sales'] = df['JP_Sales'].astype(float)
df['JP_Sales'] = df['JP_Sales']*1000000
```

```python
df['JP_Sales'].loc[t] = les_Z['Z']
df['JP_Sales'] = df['JP_Sales'].astype(int)
```

```python
 les_G = df.loc[df.Other_Sales.str.contains('K')]
 les_G['G'] = change(les_G['Other_Sales'])
 les_G['G'] = les_G['G']*1000
 r = les_G['G'].index;
 df['Other_Sales'] = df['Other_Sales'].str.replace("K", "");
 df['Other_Sales'] = df['Other_Sales'].str.replace("M", "");
 df['Other_Sales'] = df['Other_Sales'].astype(float)
 df['Other_Sales'] = df['Other_Sales']*1000000
 df['Other_Sales'].loc[r] = les_G['G']
 df['Other_Sales'] = df['Other_Sales'].astype(int)
```

```python
df['Total'] = df['NA_Sales'] + df['EU_Sales'] + df['JP_Sales'] +df['Other_Sales']
#Totla 컬럼 = 총 판매량
df = df[df.Year>1950]
(df['Year']<1950).sum()
df['Year'] = df['Year'].astype(int)
#Year의 이상치 제거 
```

```python
df_na = df.groupby('Genre')['NA_Sales'].sum()
df_na = df_na.astype(int)
df_na_pie = df_na.sort_values(ascending=False)
df_na_pie = df_na_pie.reset_index()#NA_Sales의 장르 선호도 조사 
```

```python
df_eu = df.groupby('Genre')['EU_Sales'].sum()
df_eu = df_eu.astype(int)
df_eu_pie = df_eu.sort_values(ascending=False)
df_eu_pie = df_eu_pie.reset_index()#EU_Sales 장르 선호도 조사 
```

```python
df_jp = df.groupby('Genre')['JP_Sales'].sum()
df_jp = df_jp.astype(int)
df_jp_pie = df_jp.sort_values(ascending=False)
df_jp_pie = df_jp_pie.reset_index()#JP_Sales 장르 선호도 조사 
```

```python
df_ot = df.groupby('Genre')['Other_Sales'].sum()
df_ot = df_ot.astype(int)
df_ot_pie = df_ot.sort_values(ascending=False)
df_ot_pie = df_ot_pie.reset_index()#Other_Sales 장르 선호도 조사 
```

```python
df_to = df.groupby('Genre')['Total'].sum()
df_to = df_to.astype(int)
df_to_pie = df_to.sort_values(ascending=False)
df_to_pie = df_to_pie.reset_index()#Total 장르 선호도 조사 
```
